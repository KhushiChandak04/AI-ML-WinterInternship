```python
import pandas as pd

# Load a sample dataset with 10+ columns (Electricity Load or similar dataset)
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/00374/energydata_complete.csv"
data = pd.read_csv(url)

# Inspect the dataset
print(data.head())
print(f"Columns: {data.columns}")
```

```python
pd.options.display.max_columns = None
```

```python
data
```

```python
data['date'] = pd.to_datetime(data['date'], format='%Y-%m-%d %H:%M:%S')

# Drop unnecessary columns
data = data.drop(columns=['rv1', 'rv2'])
```

```python
data = data.sort_values(by='date')
data = data.drop(columns=['date'])
```

```python
import numpy as np
# Function to create temporal splits
def create_temporal_splits(data, input_window, prediction_window):
    X, y = [], []
    for i in range(len(data) - input_window - prediction_window):
        X.append(data.iloc[i:i + input_window].values)  # Use numeric data only
        y.append(data.iloc[i + input_window:i + input_window + prediction_window, 0].values)  # Predict 'Appliances'
    return np.array(X), np.array(y)

# Generate splits
input_window = 80
prediction_window = 20
X, y = create_temporal_splits(data, input_window, prediction_window)

print(f"X shape: {X.shape}")
print(f"y shape: {y.shape}")

# Define input and prediction windows
input_window = 80
prediction_window = 20

# Create temporal splits
X, y = create_temporal_splits(data, input_window, prediction_window)
```

```python
train_size = int(len(X) * 0.8)
X_train, y_train = X[:train_size], y[:train_size]
X_test, y_test = X[train_size:], y[train_size:]

print(f"Training Data Shape: X={X_train.shape}, y={y_train.shape}")
print(f"Testing Data Shape: X={X_test.shape}, y={y_test.shape}")
```

```python
X_train
```

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import SimpleRNN, Dense

# Define the RNN model
model = Sequential([
    SimpleRNN(250, activation='tanh', input_shape=(None, X_train.shape[2])),
    Dense(1)  # Output layer (adjust size as needed for your task)
])

# Compile the model
model.compile(optimizer='adam', loss='mse', metrics=['mae'])

# Print the model summary
model.summary()


```

```python
# Train the model
history = model.fit(X_train, y_train, epochs=20, batch_size=32, validation_data=(X_test, y_test))
```
