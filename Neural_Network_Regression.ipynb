```python
#importing the lib
import tensorflow as  tf
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# loading the dataset
dataset = pd.read_csv("https://raw.githubusercontent.com/stedy/Machine-Learning-with-R-datasets/master/insurance.csv")
dataset
```

```python
#converting text to numbers
dataset_oneHot = pd.get_dummies(dataset)
dataset_oneHot.head()
```

```python
#seperate features and labels
X = dataset_oneHot.drop("charges", axis = 1)
X.head()
```

```python
y = dataset_oneHot["charges"]
y.head()
```

```python
#Splitting data  into training and test set
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20, random_state = 42)
```

```python
X_train.head()
```

```python
#Building a neural network
tf.random.set_seed(42)
```

```python
charges_model = tf.keras.Sequential()
charges_model.add(tf.keras.layers.Dense(11))
charges_model.add(tf.keras.layers.Dense(100, activation = "relu"))
charges_model.add(tf.keras.layers.Dense(50, activation = "relu"))
charges_model.add(tf.keras.layers.Dense(100, activation = "relu"))
charges_model.add(tf.keras.layers.Dense(100, activation = "relu"))
charges_model.add(tf.keras.layers.Dense(1))
```

```python
#Compile the model
charges_model.compile(loss=tf.keras.optimizers.Adam(learnin_rate=0.001),metrices=["mae"])
```
